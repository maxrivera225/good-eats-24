```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.pharmchek.com/resources/court-cases"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
states = soup.find_all(class_ ="l-listing__category")
states
```

```python
cases = soup.find_all("div", class_="case__summary")
```

```python
print(cases)
```

```python
case_summary = [summary.text.strip() for summary in cases]
print(case_summary)
```

```python
states_summary = [summary.text.strip() for summary in states]
print(states_summary)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = states_summary)
df
```

```python
column_data = soup.find_all("div", class_="case__summary")
```

```python
for row in column_data:
    row_data = soup.find_all("div", class_="case__summary")
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
